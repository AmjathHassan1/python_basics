```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('ABC Company.xlsx')
df
```

```python
df['Height'] = np.random.randint(150, 180, size=len(df))
df
```

```python
# ● Determine the distribution of employees across each team and calculate
# the percentage split relative to the total number of employees. (2 marks)

team_counts = df['Team'].value_counts()
team_counts
```

```python
team_percentage = df['Team'].value_counts(normalize=True) * 100
print("Team Percentage:", team_percentage)

plt.pie(team_percentage, labels=team_counts.index, autopct='%1.1f%%')
plt.show()
```

```python
for i in team_counts:
    print(i/df.shape[0]*100)
```

```python
# Insights--------------------------------

'''
The analysis shows that some teams have a larger share of employees compared to others, indicating uneven workforce distribution across the organization. The pie chart clearly highlights which teams make up the highest and lowest proportions of the total staff.
'''
```

```python
# ● Segregate employees based on their positions within the company. (2
# marks)

grp_by_position = df.groupby('Team')
grp_by_position
position_counts = grp_by_position['Position'].value_counts()
position_counts

```

```python
'''
Insight

Employee segregation by position shows that certain roles are more common within specific teams, indicating how responsibilities are distributed. Some positions appear frequently across teams, while others are limited to only a few teams.

'''
```

```python
# Graph Plot ----------------------------------------

df_plot = position_counts.unstack(level='Position')

plt.figure(figsize=(10, 6))

df_plot.plot(kind='bar', stacked=True, ax=plt.gca(), rot=0)

plt.title('Distribution of Positions within Each Team')
plt.xlabel('Team')
plt.ylabel('Count of Employees')
plt.legend(title='Position')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()

plt.show()
```

```python
'''
The stacked bar chart shows how different positions are distributed within each team. Some teams have a wider variety of roles, while others rely heavily on specific positions. This highlights how each team is structured based on its functional needs
'''
```

```python
# ● Identify the predominant age group among employees. (2 marks)

age_lst = [18, 25, 35, 45, 55, 65]
age_grp = ['18-25', '26-35', '36-45', '46-55', '56-65']

age_group_count = pd.cut(df['Age'], bins=age_lst, labels=age_grp, right=True).value_counts()


print("Age Group Distribution:")
print(age_group_count)

predominant_group = age_group_count.idxmax()
print("Predominant Age Group:", predominant_group)

# df
```

```python
sns.barplot(age_group_count)
```

```python
'''
The age distribution shows that employees are not spread evenly across all age groups. The analysis identifies one age range as the most dominant, indicating that the workforce is mainly concentrated in this particular age bracket.
'''
```

```python
# ● Discover which team and position have the highest salary expenditure. (2
# marks)

team_salary = df.groupby('Team')['Salary'].sum()
highest_team = team_salary.idxmax()
print("Team with Highest Salary:", highest_team)

position_salary = df.groupby('Position')['Salary'].sum()
highest_position = position_salary.idxmax()
print("Position with Highest Salary Expenditure:", highest_position)

```

```python
print(" Salary BY Team: ", team_salary)
sns.barplot(team_salary)
```

```python
print(" Salary BY Position: ", position_salary)
sns.barplot(position_salary)
```

```python
'''
The results show that one team accounts for the highest total salary expenditure, indicating it may have more employees or higher-paid roles. Similarly, one position stands out with the greatest salary cost, suggesting it is either highly valued or typically filled by senior, higher-earning employees.
'''
```

```python
# ● Investigate if there's any correlation between age and salary, and
# represent it visually. (2 marks)

df1=df[['Salary', 'Age']]
sns.heatmap(df1.corr(), annot=True)
correlation = df['Age'].corr(df['Salary'])
print("Correlation:")
print(correlation) # here in correlation is 0.21 ,ie, nearly 0 that means  co relataion is a weak +ve co rrelation
#
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
# when analizing the scatter plot when age increase there is a slight tendency to increase salary
```

```python
'''
The correlation between age and salary is weakly positive, meaning salary tends to increase slightly with age, but the relationship is not strong. The heatmap confirms that age has only a minor influence on salary
'''
```
