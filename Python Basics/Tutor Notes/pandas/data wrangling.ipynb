```python
import pandas as pd
```

```python
df1=pd.DataFrame({"name":["anu","manu"],"rollnum":[12,13]})
df1
```

```python
df2=pd.DataFrame({"name":["appu","anand"],"rollnum":[8,9]})
df2
```

```python
df=pd.concat([df1,df2])
df
```

```python
df3=pd.DataFrame({"location":["ekm","pkd"],"mark":[45,50]})
df3
```

```python
df=pd.concat([df1,df3],axis=1)
df
```

```python
df4=pd.DataFrame({"rollnum":[12,25],"age":[27,28]})
df4
```

```python
df1
```

```python
df4
```

```python
df=pd.merge(df1,df4,how="inner",on="rollnum")
df
```

```python
df=pd.merge(df1,df4,how="left",on="rollnum")#df1-left,df4-right
df
```

```python
df=pd.merge(df1,df4,how="right",on="rollnum")
df
```

```python
df=pd.merge(df1,df4,how="outer",on="rollnum")
df
```
