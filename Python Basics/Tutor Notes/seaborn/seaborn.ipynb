```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df
```

```python
#jointplot
sns.jointplot(x="size",y="total_bill",data=df)
```

```python
sns.jointplot(x="size",y="total_bill",data=df,hue="sex")
```

```python
sns.jointplot(x="size",y="total_bill",data=df,hue="day")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,hue="time")
```

```python
#pairplot
sns.pairplot(df)
```

```python
sns.countplot(x="sex",data=df,hue="smoker")
```

```python
sns.countplot(x="sex",data=df,hue="time")
```

```python
sns.histplot(df["day"],color="green")
```

```python
sns.barplot(x="total_bill",y="smoker",data=df)
```
