```python
import time
```

```python
for page in range(1,4):
    url="https://www.flipkart.com/search?q=mobile+phones&otracker-search&otracker1-search&marketplace-FLIPKART&as-show-on&as-off&page-{page}"
    driver.get(url)
    time.sleep(1)
    driver.maximize_window()
    data_full=[]
    containers=driver.find_elements(By.CSS_SELECTOR,'div[class="yKfJKb row"]')
    for i in containers:
        name=i.find_element(By.CSS_SELECTOR,'div[class="KzDlHZ"]').text
    #print(name)
        price=i.find_element(By.CSS_SELECTOR,'div[class="Nx9bqj _4b5DiR"]').text
    #print(price)
        rating=i.find_element(By.CSS_SELECTOR,'div[class="XQDdHH"]').text
    #print(rating)
        data={"name":name,"price":price,"rating":rating}
        data_full.append(data)
        print(data_full)
    page_df=pd.DataFrame(data_full)
    page_df.to_csv(output_csv,mode="a",header=False,index=False)
    time.sleep(1)
```
