```python
#t test---paired t test
```

```python
import numpy as np
from scipy import stats
```

```python
a=np.array([115 ,120 ,112 ,128 ,122 ,125, 110, 117, 119,114])
b=np.array([120, 122 ,118 ,130 ,125 ,128 ,115 ,121 ,123, 119])
```

```python
#define hypothesis
#null hypothesis(H0):the new drug has no effect on the blood pressure
#alternate hypothesis(H1):the new drug has an effect on blood pressure
```

```python
#significance leval   5/100=0.05---significance leval
alpha=0.05#0.01
```

```python
t_stat,p_val=stats.ttest_rel(a,b)
```

```python
t_stat#measures how far apart the two group means are,large t_stat value----bigger difference
```

```python
p_val#small p_value-difference is statisticlly significant
```

```python
if p_val<alpha:
  decision="reject H0"
else:
  decision="failed to reject H0"
```

```python
decision
```

```python
if decision=="reject H0":
  conclusion="there is a statistical evidence that avg bp before and after has effect"
if decision=="failed to reject H0":
  conclusion="these is insufficient evidence to claim that avg bp before and after has effect"
```

```python
conclusion
```

```python
# 2.a company claims that the average lifetime of its light bulb is 1200 hours.a
# consumer protection agency suspects that the bulbs actually lasts less than
# 1200 hours.

# H0:evidence suggests the mean lifetime is < 1200 hours
# H1:evidence suggests the mean lifetime is not less than 1200 hours


# data=np.array([1180,1195,1201,1175,1188,1190,1172,1185,1193,1200,1182,1191,1178
#  1187,1184])

# population avg=1200
```

```python
data=np.array([1180,1195,1201,1175,1188,1190,1172,1185,1193,1200,1182,1191,1178,1187,1184])
```

```python
#<1200---left t test
#>,>=    right t test
```

```python
mu=1200
```

```python
t_stat,p_val=stats.ttest_1samp(data,mu)
```

```python
t_stat
```

```python
p_val
```

```python
alpha=0.05
```

```python
if p_val<alpha:
  print("reject h0")
else:
  print("fails to reject H0: evidence suggest that the mean lifetime is <1200 hrs")
```
