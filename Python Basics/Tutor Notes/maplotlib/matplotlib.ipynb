```python
import matplotlib.pyplot as plt #submodule
```

```python
x=[5,12,8,2,1]#(5,4)
y=[4,1,9,0,3]
plt.plot(x,y)
plt.show()
```

```python
#passing values as an array
import numpy as np
y=np.array([1,5,2,7])
y
```

```python
plt.plot(y,"*:g")#marker,g-color
plt.show()
```

```python
plt.plot(y,linestyle="dotted",color="r",marker="+",linewidth=3)
```

```python
x=[150,155,160,165]
y=[40,42,45,48]
plt.xlabel("height in cm")
plt.ylabel("weight in kg")
plt.title("HEIGHT VS WEIGHT")
plt.plot(x,y)
plt.show()
```

```python
#barplot:
x=["a","b","c","d"]
y=[50,60,70,80]
plt.bar(x,y,color="green")
```

```python
plt.barh(x,y)
```

```python
#histogram:frequency of observations
x=[1,1,2,2,2,2,3,4,6,6,6,7]
plt.hist(x,align="mid",)
```

```python
#boxplot
x=[20,25,30,35,40]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40]
plt.pie(x,labels=["apple","orange","grapes","banana"],colors=["red","yellow","blue","green"])
plt.legend(loc="upper right")
plt.show()
```

```python
#area chart
x=[1,2,3,4,5]
y=[10,20,15,25,30]
plt.fill_between(x,y,color="green")
plt.plot(x,y)
plt.xlabel("month")
plt.ylabel("sales")
```

```python
#stacked barchart
subjects=["math","science","english","history"]
boys=[80,70,90,60]
girls=[85,75,95,70]
plt.bar(subjects,boys,label="boys",color="green")
plt.bar(subjects,girls,bottom=boys,label="girls",color="red")
```

```python
#bubble chART
x=[10,20,30,40,50]
y=[15,25,20,35,45]
sizes=[100,200,300,400,500]
plt.scatter(x,y,s=sizes,color="red")
```

```python
#stem plot
x=[1,2,3,4,5]
y=[5,7,9,6,3]
plt.stem(x,y)
```

```python
#sub plots
x1=[1,2,3,4]
y1=[3,7,9,2]
plt.subplot(3,1,1)#row,column,order
plt.plot(x1,y1)


x2=[3,2,5,1]
y2=[3,4,5,6]
plt.subplot(3,1,2)
plt.plot(x2,y2)


x3=[1,2,3,4]
y3=[3,7,9,2]
plt.subplot(3,1,3)#row,column,order
plt.plot(x3,y3)
```
