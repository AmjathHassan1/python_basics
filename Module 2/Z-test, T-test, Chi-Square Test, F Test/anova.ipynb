```python
# 4.a company decided 3 types of marketing campaigns to test which one brings in higher sales

# company A: Email marketing
# company B: social media ads
# company C:TV commercials
# they measure the weekly sales (in thousand of dollars) from stores exposed to each campaign

# company A(email):[12,15,14,10,13]
# comapny B(social media):[22,25,20,23,24]
# comapny C(TV ads):[30,28,35,32,31]

# the company wants to check is there a significant difference in mean sales between 3 campaigns
```

```python
#sales-dependent
```

```python
#h0:no significant difference in campaign performance
#h1:atleast one campaign led to significantly different sales in increase
```

```python
import scipy.stats as stats
```

```python
campaign_A=[12,15,14,10,13]
campaign_B=[22,25,20,23,24]
campaign_C=[30,28,35,32,31]
```

```python
f_stat,p_val=stats.f_oneway(campaign_A,campaign_B,campaign_C)
```

```python
f_stat
```

```python
p_val
```

```python
alpha=0.05
```

```python
if p_val<alpha:
  print("reject null hypothesis")
else:
  print("failed to reject null hypo")
```

```python
#two way anova
# 5.two way anova
# a company wants to study sales performance based on two factors.
# factor 1:marketing campaign type
# campaign A(Email)
# campaign B(social media)
# factor 2:region
# region1 :urban
# region2:rural
data={"sales":[15,18,20,10,12,11,25,28,30,20,22,19],
     "campaign":["A","A","A","A","A","A","B","B","B","B","B","B"],
     "region":["urban","urban","urban","rural","rural","rural","urban","urban","urban","rural","rural","rural"]}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols#ordinary least square
```

```python
model=ols("sales~C(campaign)+C(region)+C(campaign):C(region)",data=df).fit()
```

```python
anova_table=sm.stats.anova_lm(model,typ=2)#1,2,3
```

```python
#degrees of freedom=no.of categories-1
```

```python
anova_table
```
