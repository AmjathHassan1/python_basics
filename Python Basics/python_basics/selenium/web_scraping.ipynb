```python
output_csv="flipkart_data.csv"
import pandas as pd
columns=['name', 'price', 'rating']
pd.DataFrame(columns=columns).to_csv(output_csv, index=False)
```

```python
import time

for page in range(1,4):
    url=f"https://www.flipkart.com/search?q=mobile+phones&otracker-search&otracker1-search&marketplace-FLIPKART&as-show-on&as-off&page-{page}"
    driver.get(url)
    time.sleep(1)
    driver.maximize_window()
    containers=driver.find_elements(By.CSS_SELECTOR,'div[class="yKfJKb row"]')
    data_full=[]
    print("con", containers)
    for i in containers:
        try:
            name=i.find_element(By.CSS_SELECTOR,'div[class="KzDlHZ"]').text.strip()
            print("name--", name)
            price=i.find_element(By.CSS_SELECTOR,'div[class="Nx9bqj _4b5DiR"]').text.strip()
            print("price--", price)
            rating=i.find_element(By.CSS_SELECTOR,'div[class="XQDdHH"]').text
            print("rating--", rating)
        except:
            rating = None
            print("no rating")
        data= {'name':name, 'price':price, 'rating':rating}
        data_full.append(data)
        page_df=pd.DataFrame.from_records(data_full)
        time.sleep(1)
        page_df.to_csv(output_csv,mode='a', header=False, index=False)
    print(len(data_full))
```
